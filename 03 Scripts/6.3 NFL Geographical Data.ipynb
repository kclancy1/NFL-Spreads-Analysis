02. Data Cleaning/Wrangling

```python
# drop away_favorite column
spreads = spreads.drop(columns=['away_favorite'])
```

```python
# rename home_favorite column 0 = away favorite 1 = home favorite 
spreads = spreads.rename(columns={'home_favorite':'favorite'})
```

```python
spreads.head()
```

```python
# split game location column into 2 columns 'city' and 'state'
spreads[['city', 'state']] = spreads['game_location'].str.split(', ', expand =True)
```

```python
spreads.head(10)
```

```python
# now that we have city and state columns we can drop the game_location column
spreads = spreads.drop(columns=['game_location'])
```

```python
# calculate average total points scored in each stadium
avg_points_per_stadium = spreads.groupby('stadium')['total_points_scored'].mean()
```

```python
print(avg_points_per_stadium)
```

```python
avg_points_per_stadium.columns = ['stadium_name', 'avg_points_scored']
```

```python
# merge columns stadium, city, state, and geo coordinates from spreads onto new subset df
plot_data = pd.merge(avg_points_per_stadium, spreads[['stadium', 'city', 'state', 'stadium_latitude', 'stadium_longitude']], on='stadium', how='left')
```

```python
print(plot_data)
```

```python
# drop duplicate columns because we only need one row for each stadium
plot_data = plot_data[['stadium', 'total_points_scored', 'city', 'state', 'stadium_latitude', 'stadium_longitude']].drop_duplicates()
```

```python
print(plot_data)
```

```python
# fix decicmal places of the total_points_scored column
plot_data.total_points_scored = plot_data.total_points_scored.round(1)
```

```python
print(plot_data)
```

```python
# reset the index of plot_data
plot_data.reset_index(drop = True)
```

```python
# map 
map = folium.Map(location = [plot_data.stadium_latitude.mean(), plot_data.stadium_longitude.mean()], zoom_start = 4)
```

```python
map
```

```python
# map
map = folium.Map(location = [plot_data.stadium_latitude.mean(),plot_data.stadium_longitude.mean()], zoom_start=3)

for i, row in plot_data.iterrows():
    stadium_name = row['stadium']
    avg_points_scored = row['total_points_scored']
    latitude = row['stadium_latitude']
    longitude = row['stadium_longitude']

# create marker with popup
    folium.Marker([latitude, longitude], popup=(stadium_name, avg_points_scored)
).add_to(map)
```

```python
map
```

I created a marker map instead of a choropleth as I think it fits my data better. Here the icons represent the geographical location of stadiums and when clicked on they display the name of the stadium and average points scored. One thing I noticed is that the average number of total points scored in games outside of the United States are significantly lower than average points scored in games in the US. It is a small sample size but I found it interesting. Also average points scored in dome stadiums are among the highest across the league, there are many variables that could contribute to this though.
