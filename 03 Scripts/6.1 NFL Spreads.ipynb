03. Additional Cleaning

```python
# rename column spread_line to point_spread
# a point_spread refers to the margin of victory assigned by oddsmakers in any given game
# a value of -2.5 indicates that oddsmakers believe that team will win by 3 points or more

spreads.rename(columns = {'spread_line':'point_spread'}, inplace = True)
```

```python
# check output
spreads.head(10)
```

```python
# map team_id to correcct teams
spreads['team_home'] = spreads.team_home.map(teams.set_index('team_name')['team_id'].to_dict())
spreads['team_away'] = spreads.team_away.map(teams.set_index('team_name')['team_id'].to_dict())
```

```python
# check output
spreads.head(10)
```

```python
# create home and away favorite columns 
spreads.loc[spreads.team_favored_id == spreads.team_home, 'home_favorite'] = 1
spreads.loc[spreads.team_favored_id == spreads.team_away, 'away_favorite'] = 1
spreads.home_favorite.fillna(0, inplace = True)
spreads.away_favorite.fillna(0, inplace = True)
```

```python
# check output
spreads.head(10)
```

```python
# define total_line 
# 0 = under, 1 = over, 2 = push
def total_result(row):
    total_score = row['score_home'] + row['score_away']
    if total_score > row['total_line']:
        return '1'
    elif total_score < row['total_line']:
        return '0'
    else: 
      return '2'  
```

```python
# create total_result column
spreads['total_result'] = spreads.apply(total_result, axis = 1)
```

```python
# check output
spreads.head(10)
```

```python
# create game result column
# 0 = away win, 1 = home win
spreads['result'] = (spreads.score_home > spreads.score_away).astype(int)
```

```python
# check output 
spreads.head(5)
```

```python
spreads.head(25)
```

```python
# create column to determine who covered the spread
# 1 = Home, 0 = Away, 2 = Push
# a push occurs when a team won/lost by the exact number of the spread
def cover_result(row):
    if row['team_favored_id'] == row['team_home']:
        spread_result = row['point_differential'] + row['point_spread']
    else:
        spread_result = row['point_differential'] - row['point_spread']

    if spread_result > 0:
        return '1'
    elif spread_result < 0:
        return '0'
    else:
        return '2'
```

```python
spreads['cover_result'] = spreads.apply(cover_result, axis=1)
```
